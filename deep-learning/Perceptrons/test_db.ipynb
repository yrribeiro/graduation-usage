```python
from Perceptron import Perceptron

import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = np.load('datasets_np/db/X_train.npy')
y_train = np.load('datasets_np/db/y_train.npy')

X_test = np.load('datasets_np/db/X_test.npy')
y_test = np.load('datasets_np/db/y_test.npy')
```

```python
p = Perceptron()
p.fit(X_train, y_train)
y_pred = p.predict(X_test)
print(f'acurácia: {p.score_acc(y_test, y_pred)}')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.scatter(X_train[:, 0], X_train[:, 1], marker='*', c=y_train)

x0_1 = np.amin(X_train[:, 0])
x0_2 = np.amax(X_train[:, 0])

x1_1 = (-p.weights[0] * x0_1 - p.bias) / p.weights[1]
x1_2 = (-p.weights[0] * x0_2 - p.bias) / p.weights[1]

ax.plot([x0_1, x0_2], [x1_1, x1_2], 'k')

ymin = np.amin(X_train[:, 1])
ymax = np.amax(X_train[:, 1])
ax.set_ylim([ymin - 3, ymax + 3])

plt.show()
```
